```python
#setup, installs and imports
!rm -r -f * #clear the working directory
!pip install big-sleep natsort
import os, random, shutil, re
from tqdm.notebook import trange
from IPython.display import Image, display
from big_sleep import Imagine
from torchvision.utils import save_image
from pathlib import Path
from natsort import natsorted
print("Done with imports!")
# import random
# import re
# !pip install natsort
# import shutil

```

```python
TEXT = 'throwing rocks at your window at midnight in cyberpunk style' #@param {type:"string"}
SAVE_EVERY = 20 #@param {type:"number"}
SAVE_PROGRESS = True #@param {type:"boolean"}
LEARNING_RATE = 5e-2 #@param {type:"number"}
ITERATIONS = 500 #@param {type:"number"}
SEED = random.randrange(1000000) #@param {type: "number"}


model = Imagine(
    text = TEXT,
    save_every = SAVE_EVERY,
    lr = LEARNING_RATE,
    iterations = ITERATIONS,
    save_progress = SAVE_PROGRESS,
    seed = SEED
)
# current = f"./{TEXT.replace(' ', '_')}"
helpme = "try2"
current = f"./{helpme}"
try:
    os.mkdir(current)
except:
    print('exists')
replacements = {" ": "_", ",":"", "|":"--"}
filename = TEXT
for key, value in replacements.items():
    filename = filename.replace(key, value)
for epoch in trange(1, desc = 'epochs'):
    for i in trange(ITERATIONS + 1, desc = 'iteration'):
        model.train_step(epoch, i)

        if i== 0 or i % model.save_every != 0:
            continue
        

        print(filename, i)

#         save_image(image, Path(f'./{helpme}/filename-{SEED}_{i}.png'))
        
        image = Image(f'{filename}.png')
        display(image)
#         iterc = '{:0>3}'.format(i)
#         os.system(f'cp {filename}.png {helpme}/{filename}-{SEED}_{iterc}.png')
#         count +=1

        
```

```python

current_folder = f"./{filename}-{SEED}"

try:
    os.mkdir(current_folder)
except:
    print('exists')

count = 0

for img in natsorted(os.listdir('.')):
    if re.search('[0-9].png', img):

        iterc = '{:0>3}'.format(count)
        print(f'cp {img} {current_folder}/step-{iterc}.png -r')
        os.system(f'cp {img} {current_folder}/step-{iterc}.png -r')
        count += 1
        print("okay!")
```
